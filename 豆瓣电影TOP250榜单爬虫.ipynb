```python
import requests
# 导入requests请求库

url = 'https://movie.douban.com/top250?start=0&filter='
# 导入网址

myheader = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.89 Safari/537.36'}
# 导入请求头

r = requests.get(url,headers = myheader)
# 获取网页源代码

# r.text
# 查看网页源代码

from bs4 import BeautifulSoup
# 导入解析库

sp = BeautifulSoup(r.text, 'html5lib')
# 解析网页源代码；'html5lib'为解析器

tag = sp.select_one('ol.grid_view')
# 返回满足指定样式的第一个标签本身

# tag

tags = tag.select('li')
# 返回所有满足指定样式的标签本身

# tags[1]

k=1

for t in tags:      #以第一部电影标签来测试

    # t

    title = t.select_one('span.title')
    # 

    names = title.string

    # name

    name = names.replace('\xa0','')
    
#     name

    other_name = t.select_one('span.other').string.replace('\xa0','')

    message = t.select_one('div.bd').p.get_text().replace('\n','').replace(' ','').replace('\xa0','')

#     message

    rating_num = t.select_one('span.rating_num').string

    rate_mans = t.select_one('div.star').select('span')[-1].string[:-3]

    idea = t.select_one('span.inq').string

    print(k,name,other_name,message,rate_num,rate_mans,idea)
    
    k+=1
```

```python
import requests,time
from bs4 import BeautifulSoup
myheader = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0;WOW64) AppleWebKit/537.36 (KHTML, like Gecko)Chrome/66.0.3359.139 Safari/537.36'}
rank = 1 # 电影排名计数器初始化
fout = open('豆瓣电影TOP250信息.txt','w',encoding='utf-8') # 打开文件以保存数据
for index in range(0,250,25):     # 循环构造网页网址
    url = 'https://movie.douban.com/top250? start='+str(index)+'&filter='#拼接构造当前页网址
    r = requests.get(url, headers = myheader) # 请求当前页源代码
    sp = BeautifulSoup(r.text, 'html5lib') # 解析当前页
    tag = sp.select_one('ol.grid_view') # 查找25部电影的父标签
    tags = tag.select('li')    #查找25部电影标签列表
    for t in tags:      ## 遍历当前25部电影
        title = t.select_one('div.hd')
        title.a.get_text()
        name1 = ''.join(title.a.stripped_strings)    # 提取处理并保存电影名
        name = '《' + name1 + '》'  + '\n'   # 给电影名加上书名号
        body = t.select_one('div.bd')    
        info1 = ''.join(body.p.stripped_strings)    # 提取处理并保存电影信息
        info = '电影信息：' + info1  + '\n'
        body.select_one('span.rating_num')
        rating1 = body.select_one('span.rating_num').string    # 提取并保存评分
        rating = '评分：' +rating1  + '\n'
        rate_name1 = body.div.select('span')[-1].string[:-3]    # 提取并保存评论人数
        rate_name = '评论人数：' + rate_name1 + '人'  + '\n'
        body.select_one('span.inq').string
        comment = body.select_one('span.inq').string    # 查找精彩评论标签
        if tag != None: # 如果存在则提取评论
            comment = tag.string
        else: #否则评论为空
            comment = '' 
        print(name,info,rating,rate_name,comment)
        print(rank,name,info,rating,rate_name,comment,sep='\t',file=fout) # 将数据写入文件
        rank += 1 # 计数器加一
```
