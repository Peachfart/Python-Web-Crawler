```python
import requests    #导入网络数据请求库
from bs4 import BeautifulSoup    #导入网页解析库
from time import sleep     #导入暂停函数

fout = open('《三国演义》.txt','w',encoding = 'utf-8')    #新建文本文件以写入

for k in range(1,121):
    url = 'http://sdmz.org/sanguoyanyi/'+str(k)+'.html'    #设置网址
    r = requests.get(url)    #请求网页源代码
    r.encoding = 'utf-8'    #设置字符编码
    sp = BeautifulSoup(r.text,'lxml')    #解析网页源代码
    sp.find('h2')     #提取标题
    title = sp.h2.string
    t = sp.select_one('div.span9')     #查找段落的父标签
    paras = t.select('p')    #查找段落标签的列表
    paras = paras[:-2]    #切除不需要的段落
    print(title + '\n',file = fout)    #将章标题写入文件
    for p in paras:    #将章文本写入文件
        print('\u3000'*2 + p.string + '\n',file = fout)
    print(f'-------------第{k:03}回爬取完成！----------')
    sleep0(0.5)    #暂停0.5秒
    
fout.close()    #关闭并保存文件
```
