```python
# 知乎Python精华话题爬虫
from selenium import webdriver    #
from selenium.webdriver.common.keys import Keys    #
from time import sleep

browser2 = webdriver.Chrome()    #打开空白浏览器
browser2.get('https://www.zhihu.com/topic/19552832/top-answers')    #访问指定网站

page2 = browser2.find_element_by_tag_name('html')

for k in range(10):
    page2.send_keys(Keys.END)
    sleep(2)

html = browser2.page_source    #保存当前页面源代码
```

```python
from bs4 import BeautifulSoup
```

```python
sp = BeautifulSoup(html,'lxml')    #解析网页源代码
```

```python
t = sp.select_one('div.List')     #查找祖父标签
```

```python
topics  = t.div.select('div.List-item')      #查找话题标签列表
```

```python
topic = topics[0]
```

```python
for topic in topics:
    title = topic.h2.a.string

    topic_url = 'https:' + topic.h2.a.get('href')
    
    try:
        
        author = topic.select_one('span.UserLink.AuthorInfo-name').a.string
    except:
        author = '知乎用户'
    intro = topic.select_one('span.RichText').getText()
    print(title,topic_url,author,intro,sep='\t')
```
